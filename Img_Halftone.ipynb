```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import halftone as ht
import imageio
from PIL import Image
import PIL.Image
```

```python
InputIm_FileName="floare_sa_gr_fc.jpg"
InImg=cv2.imread(InputIm_FileName,0)
```

```python
H,W=InImg.shape
DPI=80               #dots per inch
plt.figure(figsize=(2.5*W/DPI+1,1*H/DPI+1))
plt.subplot(1,2,1)
plt.imshow(InImg,cmap = 'gray')
plt.subplot(1,2,2)
plt.hist(InImg.ravel(),256,[0,255])
plt.suptitle('The bintonal histogram')
plt.show()
```

```python
img = PIL.Image.open('floare_sa_gr_fc.jpg')
halftoned = ht.halftone(img, ht.euclid_dot(spacing=5, angle=0))
halftoned.save('floare_halftone.jpg')

htInputIm_FileName="floare_halftone.jpg"
htInImg=cv2.imread(htInputIm_FileName,0)


H,W=htInImg.shape
DPI=80               #dots per inch
plt.figure(figsize=(2.5*W/DPI+1,1*H/DPI+1))
plt.subplot(1,2,1)
plt.imshow(htInImg,cmap = 'gray')
plt.subplot(1,2,2)
plt.hist(htInImg.ravel(),256,[0,255])
plt.suptitle('The halftone img && histogram')
plt.show()
```
